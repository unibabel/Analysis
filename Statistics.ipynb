# Unicode CharaStatistics

```python
import pandas as pd

UCD_URL = 'http://www.unicode.org/Public/9.0.0/ucd/UnicodeData.txt'
FIELD_NAMES = '''name general_category canonical_combining_class bidi_class decomposition 
                 numeric_6 numeric_7 numeric_8 bidi_mirrored unicode_1_name iso_comment
                 mapping_uppercase mapping_lowercase mapping_titlecase'''.split()
```

```python
ucd = pd.read_table(UCD_URL, header=None, sep=';', names=FIELD_NAMES)
```
